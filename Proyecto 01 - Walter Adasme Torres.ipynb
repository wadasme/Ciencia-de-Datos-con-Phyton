#### **EJERCICIO N° 01**

Escribir un programa que pregunte al usuario por las ventas de los últimos 5 días, y muestre por pantalla una serie con los datos de las ventas por producto: Nro de unidades y precio unitario, indexada por los días, antes y después de aplicarles un descuento del 5%.

```python
# Determinamos la existencia de 2 productos, con precio unitario pre-establecido
# Productos : Arroz    Precio Unitario 3.6 por kilo
#             Azúcar   Precio Unitario 4.5 por kilo

# Importamos las librerías a usar
import numpy  as np
import pandas as pd

# Declaramos el uso de fechas (Para mostrar en pantalla)
from datetime import datetime
from datetime import timedelta

# Variable Dias, indica el número de días a cargar datos
# Variable PU,   indica el precio unitario por cada producto
# Variable Dcto, indica el descuento que se realizará a los productos.
Dias   = 5
PU_1   = 3.6
PU_2   = 4.5
Dcto   = 0.05   # ( Descuento es en decimales, p.e. 5% entonces se indica 0.05 )

# Establecemos el día inicial, hoy menos "n" días (indicado en la variable Dias)
Ini = datetime.now() - timedelta(days=Dias)

# Declaramos una lista con las fechas para cargar las fechas calculadas de menor a mayor
Lista_dias = [Ini.strftime("%d-%m-%Y")]
for i in range(1, Dias):
    Nuevo_dia = Ini + timedelta(days=i)
    Lista_dias.insert(i, Nuevo_dia.strftime("%d-%m-%Y"))

# Declaramos las listas de Cantidad y Ventas, uno por cada producto
# Se solicita el ingreso de las ventas de los últimos "n" días (indicado en la variable Dias) por producto
Cantidad_producto_1 = []
Cantidad_producto_2 = []
Ventas_producto_1   = []
Ventas_producto_2   = []
for j in range(0, Dias):
    print("Ventas del día : ", Lista_dias[j])
    V1 = float(input('Ingrese la venta de Arroz (cantidad)  : '))
    V2 = float(input('Ingrese la venta de Azúcar (cantidad) : '))
    Cantidad_producto_1.insert(j, V1)
    Cantidad_producto_2.insert(j, V2)
    Ventas_producto_1.insert(j, V1*PU_1)
    Ventas_producto_2.insert(j, V2*PU_2)
    print("\n")

# Convertimos a Array, y luego a Dataframe transponiendo el array 
#      tanto para ventas por cantidad como también para ventas totales, agregando la fechas en el índice y la descripción de productos en cabecera

# Cantidad
PU_1_c = "Arroz  ( Cant )  "
PU_2_c = "Azúcar ( Cant )  "
Lista_Columnas_Cantidad = [ PU_1_c, PU_2_c ]

Lista_Total_Cantidad = Cantidad_producto_1 + Cantidad_producto_2
Array_Cantidad = np.array(Lista_Total_Cantidad).reshape(2, Dias)
DF_Cantidad = pd.DataFrame( data=Array_Cantidad.T, index=pd.date_range(Ini.strftime("%d-%m-%Y"), periods=Dias), columns=Lista_Columnas_Cantidad )

# Ventas
PU_1_v = "Arroz  (PU = "+str(PU_1)+")"
PU_2_v = "Azúcar (PU = "+str(PU_2)+")"
Lista_Columnas_Ventas = [ PU_1_v, PU_2_v ]

Lista_Total_Ventas = Ventas_producto_1 + Ventas_producto_2
Array_Ventas = np.array(Lista_Total_Ventas).reshape(2, Dias)
DF_Ventas = pd.DataFrame( data=Array_Ventas.T, index=pd.date_range(Ini.strftime("%d-%m-%Y"), periods=Dias), columns=Lista_Columnas_Ventas )

# Se imprime la venta de todos los días por cantidad
print(" Cantidad de productos vendidos")
print(DF_Cantidad)
print("\n")76

# Se imprime la venta de todos los días por total
print(" Total de ventas realizadas")
print(DF_Ventas)
print("\n")

# Se imprime la venta de todos los días por total y descuento aplicado
print(" Total de ventas realizadas aplicando descuento de ", Dcto*100,"%")
print(DF_Ventas*(1-Dcto))
print("\n")
```

#### **EJERCICIO N° 02**

Escribir un programa que reciba un diccionario con los ingresos mensuales de los clientes de una empresa, y devuelva una serie con el ingreso mínimo, máximo y media.

```python
# Importamos las librerías a usar
import numpy  as np

# Variable Clientes, indica la cantidad de clientes a ingresar datos
Clientes = 3

# Creamos un diccionario vacío, y una lista vacía
Ingresos_mensuales = {}
Lista_IM = []

# Iteramos la cantidad de clientes para ingresar id del cliente e ingreso mensual
for i in range(0, Clientes):
    ID = int(input("Ingrese ID del cliente     : "))
    IM = float(input("Indique el ingreso mensual : "))
    Ingresos_mensuales[ID] = IM
print("\n")
print("Diccionario ingresado")
print(Ingresos_mensuales)
print("\n")

# Iteramos el diccionario para obtener una lista de Ingresos Mensuales
jj = 0
for j in Ingresos_mensuales:
    Lista_IM.insert(jj,Ingresos_mensuales[j])
    jj += 1

# Imprimimos de la lista obtenida y del Ingreso Mensual, el valor mínimo, máximo y la media.
print("Ingreso mínimo : ", np.min(Lista_IM))
print("Ingreso máximo : ", np.max(Lista_IM))
print("Media          : ", np.mean(Lista_IM)) 
```

#### **EJERCICIO N° 03**

Escribir programa que genere y muestre por pantalla un DataFrame con los datos de la tabla siguiente:

Tabla Ejercicio 3.JPG


A continuación, generar nuevas columnas:
1.	Columna 1 =  Gasto/Ventas*100.
2.	Columna 2 =  Si Ventas>= 30000, imprimir “Meta Superada”, caso contrario imprimir “Meta no Superada”.
3.	Columna 3 =  Si Gasto>Ventas, calcular el sobre gasto.
4.	Obtener el total de las ventas de los últimos 3 meses
5.	Obtener el total de gastos de febrero a abril


```python
# Importamos las librerías a usar
import pandas as pd
import numpy  as np

# Generamos e imprimimos la tabla predeterminada
print("Tabla predeterminada")
Tabla_datos = {'Mes':['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo'], 
               'Ventas':[30500, 35600, 28300, 33900, 42500], 
               'Gastos':[22000, 23450, 18100, 35700, 32450]}
DF_datos = pd.DataFrame(Tabla_datos)
print(DF_datos)
print("\n")

# Se agrega e imprime columna "Gastos_Ventas" con operación Gastos/Ventas*100
print("Se agrega la columna \"Gastos_Ventas\", con fórmula \"Gastos/Ventas*100\"")
DF_datos["Gastos_Ventas"] = DF_datos.Gastos / DF_datos.Ventas * 100
print(DF_datos)
print("\n")

# Se agrega e imprime columna "Meta" con operación si Ventas es mayor que 30000 indique "Meta Superada", caso contrario "Meta No Superada"
print("Se agrega la columna \"Meta\", cuando \"Ventas\" es mayor a 30000 la \"Meta es Superada\", caso contrario la \"Meta No es Superada\"")
DF_datos["Meta"] = np.where(DF_datos['Ventas']>=30000, 'Meta Superada', 'Meta No Superada')
print(DF_datos)
print("\n")

# Se agrega e imprime columna "Sobre_Gasto" con operación si Gasto es mayor a Ventas indique Gastos - Ventas
print("Se agrega la columna \"Sobre_Gasto\", cuando el \"Gasto\" es mayor a la \"Venta\" se calcula el sobre gasto, \"Gastos-Ventas\"")
DF_datos["Sobre_Gasto"] = np.where(DF_datos['Gastos']>DF_datos['Ventas'], DF_datos.Gastos - DF_datos.Ventas, 0)
print(DF_datos)
print("\n")

# Total de Ventas de los últimos 3 meses
print("Se obtiene las tres últimas filas, y sobre ella se calcula el total de \"Ventas\"")
print(DF_datos.tail(3).loc[:, ["Mes", "Ventas"]])
Ventas = DF_datos.tail(3)['Ventas'].sum()
print("Venta total de los últimos 3 meses :",Ventas)
print("\n")
         
# Total de Gastos de Febrero a Abril
print("Se obtiene las filas de Febrero a Abril y sobre ella se calcula el total de \"Gastos\"")
print(DF_datos.loc[1:3].loc[:, ["Mes", "Gastos"]])
Gastos = DF_datos.loc[1:3]['Gastos'].sum()
print("Gasto total de los meses de Febrero a Abril :",Gastos)
print("\n")
```
